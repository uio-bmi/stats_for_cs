<span style="color: green;">**Possible solution:**</span>

Hypothesis test procedure:
1. Formulate the **null** and **alternative** hypothesises $H_0$ and $H_a$  
2. Choose the **threshold** for statistical **significance** (i.e. a level $\alpha$)  
3. Choose a test **statistic** with a known sampling **distribution** under $H_0$  
4. Compute the **$p$-value**, which measures how **extreme or unusual** the observed statistic is under $H_0$  
5. Draw a conclusion based on the **comparison** between the $p$-value and $\alpha$

Evidence from a hypothesis test:
The $p$-value represents the probability of observing a test statistic as extreme or more extreme than the one we observed, assuming the null hypothesis $H_0$ is true and that the assumed distribution of the test statistic is valid.

- If $p \leq \alpha$, we have sufficient evidence to reject $H_0$
- If $p > \alpha$, we do not have sufficient evidence to reject $H_0$

Type I vs Type II errors:
- Type I error: Rejecting $H_0$ when it is actually true.
- Type II error: Failing to reject $H_0$ when $H_a$ is true.

```python
def permutation_test_mean(x, y, n_perm=10000, random_state=None):
    """
    Permutation test for two independent samples x, y.
    Test statistic: difference in sample means (mean(x) - mean(y)).
    Two-sided test under the null of same distribution.
    """
    rng = np.random.default_rng(random_state)

    x = np.asarray(x)
    y = np.asarray(y)

    # Observed statistic
    obs_stat = x.mean() - y.mean()

    # Pool data
    pooled = np.concatenate([x, y])
    n_x = len(x)

    # Permutations
    perm_stats = np.empty(n_perm)
    for i in range(n_perm):
        rng.shuffle(pooled)                 # in-place shuffle
        x_perm = pooled[:n_x]
        y_perm = pooled[n_x:]
        perm_stats[i] = x_perm.mean() - y_perm.mean()

    # Two-sided p-value
    p_value = np.mean(np.abs(perm_stats) >= np.abs(obs_stat))
    return obs_stat, p_value
```

```python
def permutation_test(test_statistic, x, y, n_perm=10000, random_state=None):
    """
    Permutation test for two independent samples x, y.
    Test statistic: difference in sample means or medians.
    Two-sided test under the null of same distribution.
    """
    assert test_statistic in ("mean", "median"), "test_statistic must be 'mean' or 'median'"

    rng = np.random.default_rng(random_state)

    x = np.asarray(x)
    y = np.asarray(y)

    # Observed statistic
    if test_statistic == "mean":
        obs_stat = x.mean() - y.mean()
    elif test_statistic == "median":
        obs_stat = np.median(x) - np.median(y)
    else:
        raise ValueError("test_statistic must be 'mean' or 'median'")

    # Pool data
    pooled = np.concatenate([x, y])
    n_x = len(x)

    # Permutations
    perm_stats = np.empty(n_perm)
    for i in range(n_perm):
        rng.shuffle(pooled)                 # in-place shuffle
        x_perm = pooled[:n_x]
        y_perm = pooled[n_x:]
        if test_statistic == "mean":
            perm_stats[i] = x_perm.mean() - y_perm.mean()
        elif test_statistic == "median":
            perm_stats[i] = np.median(x_perm) - np.median(y_perm)

    # Two-sided p-value
    p_value = np.mean(np.abs(perm_stats) >= np.abs(obs_stat))
    return obs_stat, p_value
```

#### e)

Now, the t-test can also be used for our problem. Determine which of the t-tests' variance structures we have talked about that you will opt for given the plot from 1b), and identify all entities in the two t-test formulas below:

**Pooled variance:**
$$t = \frac{\bar{x} - \bar{y}}{s_p \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}$$
where $s_p = \sqrt{\frac{(n_1-1)s_x^2 + (n_2-1)s_y^2}{n_1 + n_2 - 2}}$

**Welch's:**
$$t = \frac{\bar{x} - \bar{y}}{\sqrt{\frac{s_x^2}{n_1} + \frac{s_y^2}{n_2}}}$$

Also, what are the degrees of freedoms in the two different scenarios?

<span style="color: green;">**Possible solution:**</span>

**Identify:** $\bar{x}, \bar{y}$ = sample means; $s_x^2, s_y^2$ = sample variances; $n_1, n_2$ = sample sizes; $s_p$ = pooled standard deviation.

**Degrees of freedom:**

- Equal-variance t-test:

$$df = n_1 + n_2 - 2$$

- Welch's t-test:
$$df = \frac{\left(\frac{s_x^2}{n_1} + \frac{s_y^2}{n_2}\right)^2}{\frac{(s_x^2/n_1)^2}{n_1-1} + \frac{(s_y^2/n_2)^2}{n_2-1}}$$

Compute both the equal-variance and Welch's t-tests for comparing the two groups' means, using the [`stats.ttest_ind`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html) function, as well as printing the sample variances and the ratio between them for comparison.
```python
# Equal-variance (pooled)
t_equal, p_equal = stats.ttest_ind(group_A, group_B, equal_var=True)
print(f"Equal-variance t-test: t = {t_equal:.4f}, p-value = {p_equal:.4f}")

# Unequal-variance
t_welch, p_welch = stats.ttest_ind(group_A, group_B, equal_var=False)
print(f"Welch t-test:          t = {t_welch:.4f}, p-value = {p_welch:.4f}")

# Compare variances
var_A = np.var(group_A, ddof=1)
var_B = np.var(group_B, ddof=1)
print(f"\nSample variance Group A: {var_A:.4f}")
print(f"Sample variance Group B: {var_B:.4f}")
print(f"The ratio of our variances becomes: {var_A / var_B:.4f}")
```

We also reject here, for the $\alpha = 0.05$. We should probably also use Welch's test because of different variances, if we look at a ratio of ≈ 1.5.
