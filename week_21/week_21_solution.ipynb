1e. Consider a process where the outcome of a trial depends on the previous one. Specifically, if the current state is 0, the next state is 1 with probability 0.8. If the current state is 1, the next state is 0 with probability 0.8. Write a Python function to simulate this non-independent sequence:
```python
def simulate_dependent_sequence(n_steps, initial_state=0):
    """
    Simulates n_steps of the process described above.
    Returns a list of integers (0s and 1s).
    """
    ...
```

```python
################ SOLUTION ################
import random
def simulate_dependent_sequence(n_steps, initial_state=0):
    sequence = [initial_state]
    current_state = initial_state
    
    for _ in range(n_steps - 1):
        if random.random() < 0.8:
            current_state = 1 - current_state
        sequence.append(current_state)
    return sequence
###########################################
```

<span style="color: red;">
When working with continuous random variables, the values on the probability density curve do not represent the probabilities of obtaining a specific value. <span style="color: red;">
The difference between the empirical mean and the mathematical expectation is that the former is based on an actual sample and is a statistic calculated from the values obtained, whereas the latter is a theoretical value based on mathematical results. $$\text{Var}(2\times X) = 2^2\times \text{Var}(X) = 4\times\text{Var}(X)$$

# Distributions

3a. A data scientist is working with a process that generates data following an Exponential distribution. She suggests: "If I increase my sample size $n$ to 10,000, the histogram of these individual data points will change shape from an exponential curve to a normal bell curve because of the Central Limit Theorem."Explain the misunderstanding here. Specifically, distinguish between the distribution of the underlying data and the distribution of the sample mean.

<span style="color: red;">

The central limit theorem does not apply to a sample generation process—such as the exponential distribution in this case—but rather when we take population means. If the generating process follows an exponential distribution, the more observations we have, the better we can estimate the parameters describing this exponential distribution; however, it will not become normal no matter how much we increase the number of samples.

A good example of when the central limit theorem might be used is if we assume that we can take observations every day, and different days are comparable with one another. If, instead of using the values of each observation, we used the mean for each day, we would see that as more days pass, the means would behave like a normal distribution.

</span>

4a. We are interested in finding the expected value of the transformation $Y = \cos(U) + \sin^2(U)$, where $U$ is a random variable distributed uniformly between 0 and 1.Write a Python function with the following signature:
```python
import random
import math

def monte_carlo_transformation(n_simulations):
    """
    Simulates n_simulations draws from a Uniform(0,1) distribution,
    applies the transformation cos(u) + sin^2(u) to each, 
    and returns the average of the results.
    """
    ...
```

```python
################ SOLUTION ################
import random
import math
def monte_carlo_transformation(n_simulations):
    total_sum = 0.0
    for _ in range(n_simulations):
        u = random.random()
        transformed_value = math.cos(u) + (math.sin(u) ** 2)
        total_sum += transformed_value
    return total_sum / n_simulations
###########################################
```

# Bootstrapping

7a. A junior data scientist has a dataset of 50 observations. When we use bootstrapping, we are not generating new samples; rather, we are sampling our initial observations with replacement, generating sets of populations of the same size as the original but based on the same observations we had from the outset. 7d. Write a Python function with the following signature:
```python
import random

def bootstrap_median_ci(data, n_resamples=1000):
    """
    1. Generates n_resamples of the same size as 'data' using replacement.
    2. Calculates the median for each resample.
    3. Returns a tuple (lower, upper) representing the 2.5th and 97.5th 
       percentiles of the bootstrap medians (a 95% CI).
    """
    ...
```

```python
################ SOLUTION ################
import numpy as np
def bootstrap_median_ci(data, n_resamples=1000):
    data = np.array(data)
    n = len(data)
    resamples = np.random.choice(data, size=(n_resamples, n), replace=True)
    medians = np.median(resamples, axis=1)
    lower_bound = np.percentile(medians, 2.5)
    upper_bound = np.percentile(medians, 97.5)
    return lower_bound, upper_bound
##########################################
```

7e. We want to estimate the probability that the population variance is less than 20. Write a Python function that uses bootstrapping to estimate this probability:

```python
import numpy as np

def bootstrap_variance_probability(data, threshold=20, n_resamples=1000):
    """
    Generates n_resamples from 'data'.
    Calculates the variance for each resample.
    Returns the proportion of bootstrap variances that are 
    strictly less than 'threshold'.
    """
    ...
```

```python
################ SOLUTION ################
import numpy as np
def bootstrap_variance_probability(data, threshold=20, n_resamples=1000):
    resamples = np.random.choice(data, size=(n_resamples, len(data)), replace=True)
    bootstrap_variances = np.var(resamples, axis=1, ddof=1)
    strictly_less_count = np.sum(bootstrap_variances < threshold)
    return float(strictly_less_count / n_resamples)
##########################################
```

# Hypothesis Testing

8a. A researcher conducts a t-test and obtains a p-value of 0.03. In their report, they write: "There is a 3% probability that the null hypothesis ($H_0$) is true, and therefore a 97% probability that my alternative hypothesis ($H_a$) is correct." What is the misconception? Explain what the p-value means.

<span style="color: red;">

The p-value does not measure how likely one hypothesis is compared to another. Instead, assuming the null hypothesis is true, it measures how likely it is to obtain data as extreme as or more extreme than what we have in our sample. Therefore, the p-value can be expressed as

$$P(T \ge t \mid H_0)$$

This is why having a very small p-value means that, given the null hypothesis, the probability of the data (or more extreme data) is very low; consequently, we have evidence to reject this hypothesis."

</span>

8b. Two students are debating the philosophy of hypothesis testing. Student A says: "We should treat the null hypothesis and the alternative hypothesis as equally likely from the start, just like a 50/50 coin flip, and then see which one the data supports more."Student B disagrees, arguing that the two hypotheses are not treated symmetrically. Who is right? Why?

<span style="color: red;">
The second student is right in saying that the two hypotheses do not start with the same credibility. The truth is that when we perform a hypothesis test, the starting point is to assume the null hypothesis is true. Then, we conduct the experiment and measure how strange the data are given the null hypothesis. If we see that it does not align with the observations, that is when we can reject this initial hypothesis. However, both hypotheses are not treated equally: at no point do we start by assuming the alternative hypothesis is true; rather, it is something we must prove.
</span>

8c. A pharmaceutical company is testing a new drug. They find a p-value of 0.04. The lead scientist says: "If we set our significance level ($\alpha$) at 0.05, our results are significant. However, if we set it at 0.01, they are not. Therefore, we should ignore the 0.01 threshold because it is 'inconvenient' and would make our drug look ineffective." Discuss what is the problem of this statement.

<span style="color: red;">

The procedure carried out by the company is highly discouraged and borders on what could be termed malpractice in statistical experimentation.

When conducting experiments, the correct order of the process begins by establishing the significance level ($\alpha$) that will be tolerated to evaluate whether a result is statistically significant. This is not a value to be determined at one's convenience; it must be pre-established before the testing process begins.

Furthermore, the difference between using an alpha of 0.05 versus 0.01 is that a level of 0.01 is more rigorous in avoiding the rejection of the null hypothesis when it is actually true (Type I Error). Thus, it requires much stronger evidence to reject the null hypothesis. On the other hand, a significance level of 0.05 is less strict in this regard, but a more stringent level (like 0.01) also increases the likelihood of failing to reject a false null hypothesis (Type II Error).

Therefore, one must understand the implications of these thresholds and follow the proper procedural order: limits must be set beforehand, and never after seeing the results."

</span>

# Linear and Logistic Regression

9a. A student is transitioning a problem from Linear Regression to Logistic Regression. They have to predict if a student will pass a subjector not based in a set of variables. They observe that the coefficient for "Hours Studied" ($X_1$) is $\beta_1 = 0.8$. They conclude: "This means that for every additional hour studied, the predicted probability of passing the exam increases by exactly 0.8 (or 80%)."Explain why this additive logic, which works for Linear Regression, is incorrect for Logistic Regression. Briefly describe how a change in $X$ actually affects the prediction in a logit model.

<span style="color: red;">
The student is confusing the interpretation of parameter values in Linear Regression, where this logic would be correct, with their interpretation in Logistic Regression. In the latter, the parameter value does not indicate how much the target variable changes linearly as the associated variable increases by one unit; instead, it shows how the log-odds change:

$$\ln\left(\frac{p}{1-p}\right) = \beta_0 + \beta_1 X_1 + ...$$

Because Logistic Regression applies a sigmoid function to the output of the linear combination of parameters to map the range from $[-\infty, \infty]$ to $[0, 1]$, the additive interpretation no longer works for the probability. Instead, a one-unit increase in $X_1$ increases the log-odds by $\beta_1$ (meaning the odds are multiplied by $e^{\beta_1}$), and its actual effect on the predicted probability depends on the starting value of $X$ due to the non-linear S-curve.

</span>

9b. A data scientist is building a model to detect credit card fraud. They state: "By definition, any Logistic Regression model must use a threshold of 0.5; if the predicted probability is 0.51, it is fraud; if it is 0.49, it is not." Is this true?

<span style="color: red;">
We have seen in some of the exercises from recent chapters that the threshold we establish when performing a logistic regression to classify an observation into one class or another is flexible. It will directly affect the model's sensitivity and specificity, and depending on what we want to optimize, we will favor a higher or lower threshold. We saw how the ROC curve displays the different values the threshold can take, and how specificity and sensitivity vary based on this value. In fact, this is a crucial factor to keep in mind when our classes are highly imbalanced; in such cases, it is very likely that we will not want a threshold at 0.5, but rather at a point that best adapts to the needs of our data.
</span>

9c. A dataset includes a variable "City" coded as: $1$ for New York, $2$ for Tokyo, and $3$ for London. A researcher includes this column as a numeric variable in a regression model. Why is that a mistake?

<span style="color: red;">
If we have an unordered categorical variable—as in this case, where each number corresponds to a city—using numerical values implies an inherent order among the different options. This leads the model to assume that 1 and 2 are closer to each other than, for example, 1 and 9. However, in a variable like this, that kind of order is irrelevant and makes no sense.

That is why, instead of using raw numerical variables, what we usually do is apply one-hot encoding (or creating dummy variables). In this process, a new variable is created for each category that the categorical variable can take, minus one, which serves as the reference category. These encoded variables are binary and will only 'activate' (take the value of 1) when that specific category is present. This way, we assign an independent value to each category, which aids in model interpretation and prevents the algorithm from finding spurious correlations or non-existent order.

</span>

9d. "The more variables I include in my regression model, the better it will be at predicting the future." Briefly explain why this 'maximalist' approach is dangerous.

<span style="color: red;">

Including more variables will not necessarily improve your model's predictive power.

A model trained with too many predictor variables increases in complexity—often unnecessarily—and can lead to overfitting. This occurs when the model is excellent at predicting the training data but fails to generalize well to unseen data.

For instance, consider the extreme case where a linear model has as many variables as observations ($N = K$). Such a model would achieve a training error of zero because it could perfectly fit every single data point in the training set; however, it is highly unlikely that this model could successfully generalize to new data. Furthermore, some included variables might just be pure white noise that contributes no real information. Even with irrelevant variables like these, a linear model will still assign a value to their coefficients, adding noise and destabilizing future predictions.

</span>

9e. A researcher is worried: "I cannot use Linear Regression on this dataset because the independent variable $X$ (Age) and the dependent variable $Y$ (Income) are not normally distributed; their histograms are skewed." Correct this misunderstanding.

<span style="color: red;">
There is a very common misconception that Linear Regression requires the variables themselves to be normally distributed. However, this is not a requirement; rather, it is the model's residuals (errors) that must be normally distributed to ensure that statistical inferences are valid.

The core assumptions that a linear model must satisfy are linearity, homoscedasticity (constant variance of the errors), and the normality of the residuals. There is absolutely no requirement for the independent variable ($X$) or the dependent variable ($Y$) to follow a symmetrical or normal distribution. Therefore, the researcher can safely use Linear Regression regardless of the skewness in the histograms of Age and Income.
</span>
